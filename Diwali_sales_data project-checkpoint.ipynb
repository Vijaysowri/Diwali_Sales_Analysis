```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ax = sns.countplot(x = 'Age Group',data = df1)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Age Group',data = df1, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_group_sales = df1.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
age_group_sales 
```

```python
ax2 = sns.barplot(x = 'Age Group' , y = 'Amount' , data = age_group_sales)
```

```python
ax3 = df1.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(20)
sns.set(rc = {'figure.figsize' : (20,15)})
sns.barplot(x = 'State', y = 'Orders', data = ax3)
```

```python
state_amt = df1.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(15)
```

```python
state_amt
```

```python
sns.set(rc = {'figure.figsize' : (20,15)})
ax4 = sns.barplot(x = 'State' , y = 'Amount' , data = state_amt)
```

```python
ax5 = df1.groupby(['Marital_Status'],as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False)
sns.set(rc = {'figure.figsize': (7,5)})
ax6 = sns.barplot(x = 'Marital_Status' , y = 'Orders', data = ax5)
for bars in ax6.containers:
    ax6.bar_label(bars)
```
